# Import Libraies & Data

```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
from IPython.display import clear_output
%matplotlib inline

warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("students.csv")
```

#  Understanding Data

```python
df.shape
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

Missing values present, which account for 299 rows.

<h1 style="font-size: 14px; border-radius: 20px; padding: 12px; border: 4px solid darkgreen; line-height: 1.6;">Removing the rows containing missing values:</h1>

```python
df_na = df.dropna()
print(df_na.shape)
df_na.isnull().sum()
```

```python
df_na.describe()
```

<h1 style="font-size: 14px; border-radius: 20px; padding: 12px; border: 4px solid darkgreen; line-height: 1.6;">
There are values exceeding 100 in the test scores. As these are outliers.</h1>

<h1 style="font-size: 14px; border-radius: 20px; padding: 12px; border: 4px solid darkgreen; line-height: 1.6;">
Removing the Unnecessary Data:</h1>

```python
df_cleaned = df_na[df_na['Exam_Score'] <= 100]
df_cleaned.describe()
```

<h1 style="font-size: 14px; border-radius: 20px; padding: 12px; border: 4px solid darkgreen; line-height: 1.6;">
Next,checking the data type of each column and separate them into numerical and categorical variables based on their data types.</h1>

```python
df_cleaned.info()
```

```python
df_cleaned.head()
```

```python
num_col = ['Hours_Studied', 'Attendance', 'Previous_Scores']
cat_col = ['Parental_Involvement', 'Access_to_Resources', 'Sleep_Hours', 'Extracurricular_Activities', 'Motivation_Level', 'Internet_Access', 'Tutoring_Sessions', 'Family_Income', 'Teacher_Quality', 'School_Type', 'Peer_Influence', 'Physical_Activity', 'Learning_Disabilities', 'Parental_Education_Level', 'Distance_from_Home', 'Gender']
target = 'Exam_Score'
```

# 

# Visualization

<h1 style="font-size: 14px; border-radius: 20px; padding: 12px; border: 4px solid olive; line-height: 1.6;">First implement code that prevents warnings about infinite values during visualization.</h1>

```python
df_cleaned.replace([np.inf, -np.inf], np.nan, inplace=True)
df_cleaned.isnull().sum()
```

<h1 style="font-size: 14px; border-radius: 20px; padding: 12px; border: 4px solid olive; line-height: 1.6;">
Distribution of target variable(exam_score)</h1>

```python
plt.figure(figsize=(15, 6))
sns.countplot(x=df_cleaned[target], palette='Greens')

plt.title("The distribution of target variable 'exam_score'.")
plt.grid(True, which='major', axis='y', linestyle='--', linewidth=0.7)
```

```python
a = len(df_cleaned[df_cleaned['Exam_Score'] <= 64]) / len(df_cleaned['Exam_Score']) * 100
b = len(df_cleaned[df_cleaned['Exam_Score'] >= 70]) / len(df_cleaned['Exam_Score']) * 100
c = len(df_cleaned[(df_cleaned['Exam_Score'] >= 65) & (df_cleaned['Exam_Score'] <= 69)]) / len(df_cleaned['Exam_Score']) * 100

print('The bottom', f'{a:.2f}', '% are under 64')
print('The top', f'{b:.2f}', '% are over 70')
print('The middle', f'{c:.2f}', '% are between 65 and 69')
```

<h1 style="font-size: 14px; border-radius: 20px; padding: 12px; border: 4px solid darkgreen; line-height: 1.6;">
    The bottom 21.89% of scores are under 64, while the top 24.78% scores are over 70, 50% of the scores are concentrated between 65 and 69. </h1>

# 

<h1 style="font-size: 14px; border-radius: 20px; padding: 12px; border: 4px solid olive; line-height: 1.6;">
Distribution of numeric variables</h1>

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))  

axes = axes.ravel()

for idx, col in enumerate(num_col):
    sns.histplot(x=df_cleaned[col], ax=axes[idx], color='lightgreen')
    axes[idx].grid(True, which='major', axis='y', linestyle='--', linewidth=0.7)
    axes[idx].set_title(f'Distribution of {col}')  
    
    
plt.tight_layout() 
```

<h1 style="font-size: 14px; border-radius: 20px; padding: 12px; border: 4px solid olive; line-height: 1.6;">
Relationship between the numerical variables and the target variable.</h1>

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))  

axes = axes.ravel()

for idx, col in enumerate(num_col):
    sns.scatterplot(x=df_na[col], y=df_na[target], ax=axes[idx], color='yellowgreen')  
    axes[idx].set_title(f'Relationship b/t {col} & {target}')
    plt.xlabel(col)
    plt.ylabel(target)

plt.tight_layout()  
```

```python
corr_col = ['Hours_Studied', 'Attendance', 'Previous_Scores', 'Exam_Score']
df_corr = df_cleaned[corr_col]

corr = df_corr.corr()

plt.figure(figsize=(5,4))
sns.heatmap(corr, annot=True, cmap='viridis')
```

<h1 style="font-size: 14px; border-radius: 20px; padding: 12px; border: 4px solid olive; line-height: 1.6;">
Relationship between the numerical variables and the target variable with Extracurricular Activities.</h1>

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 5))  

axes = axes.ravel()

for idx, col in enumerate(num_col):
    sns.scatterplot(x=df_na[col], y=df_na[target], ax=axes[idx], palette='PRGn', hue=df_cleaned['Extracurricular_Activities'])  
    axes[idx].set_title(f'Relationship b/t {col} & {target}')
    axes[idx].legend(loc='upper center', bbox_to_anchor=(0.5, 1.25), ncol=2, title='Extracurricular_Activities')
    plt.xlabel(col)
    plt.ylabel(target)

plt.tight_layout()  
```

# 

<h1 style="font-size: 14px; border-radius: 20px; padding: 12px; border: 4px solid olive; line-height: 1.6;">
Distribution of categorical variables</h1>

```python
fig, axes = plt.subplots(8, 2, figsize=(15, 32)) 

axes = axes.ravel()

for idx, col in enumerate(cat_col):
    sns.countplot(x=df_na[col], ax=axes[idx], palette='Greens') 
    axes[idx].set_title(f'Histogram of {col}')

plt.tight_layout()  
```

# 

<h1 style="font-size: 14px; border-radius: 20px; padding: 12px; border: 4px solid olive; line-height: 1.6;">
Relationship between the categorical variables and the target variable.</h1>

```python
fig, axes = plt.subplots(8, 2, figsize=(15, 60))  

axes = axes.ravel()

for idx, col in enumerate(cat_col):
    sns.boxplot(x=df_na[col], y=df_na[target], ax=axes[idx], palette='Greens')  
    axes[idx].set_title(f'Relationship between {col} and {target}')
    plt.xlabel(col)
    plt.ylabel(target)

plt.tight_layout()  
```
